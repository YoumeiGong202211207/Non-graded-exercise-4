```python
import random 
import time

# Store all the cards in this big deck using a suitable data structure
card_deck = [{'Name':'Diablo','Health':100,'Attack':90,'Defense':60},
{'Name':'Medusa','Health':100,'Attack':70,'Defense':70},
{'Name':'Jester','Health':120,'Attack':60,'Defense':90,},
{'Name':'Troll','Health':150,'Attack':40,'Defense':94,},
{'Name':'Specter','Health':100,'Attack':70,'Defense':70,},
{'Name':'Mist','Health':100,'Attack':75,'Defense':65,},
{'Name':'Savage','Health':100,'Attack':90,'Defense':50,},
{'Name':'Marauder','Health':100,'Attack':85,'Defense':50,},
{'Name':'Wimp','Health':110,'Attack':40,'Defense':85,},
{'Name':'Sorcerer','Health':100,'Attack':70,'Defense':55,}]


# This stores the number of cards each player will be given in the game. It can be any number, as long as it's less than or qual to the size of the card_deck
deck_size = 5

# The following is used to draw randomly n number of cards from the deck, where n = deck_size
player_deck = random.sample(card_deck, deck_size)
opponent_deck = random.sample(card_deck, deck_size)

def print_card(card):
    print(f"Name: {card['Name']}, Health: {card['Health']}, Attack: {card['Attack']}, Defense: {card['Defense']}")

def player_turn():
    print("\nYour Deck:")
    for idx, card in enumerate(player_deck):
        print(f"{idx + 1}. ", end="")
        print_card(card)

    opponent_card = random.choice(opponent_deck)
    print("\nOpponent's Card:")
    print_card(opponent_card)

    choice = int(input("\nChoose a card to play (1-5): ")) - 1
    player_card = player_deck[choice]

    if player_card['Attack'] > opponent_card['Health']:
        print(f"\nYou defeated {opponent_card['Name']}!")
        opponent_deck.remove(opponent_card)
    else:
        damage = player_card['Attack'] - opponent_card['Defense']
        if damage > 0:
            opponent_card['Health'] -= damage
            opponent_card['Defense'] = 0
        else:
            opponent_card['Defense'] -= player_card['Attack']

        print(f"\nAfter your attack:")
        print_card(opponent_card)

    time.sleep(3)

def opponent_turn():
    if not opponent_deck:
        return

    print("\nOpponent's Turn...")

    opponent_card = random.choice(opponent_deck)
    print("\nOpponent chooses:")
    print_card(opponent_card)

    print("\nYour Deck:")
    for idx, card in enumerate(player_deck):
        print(f"{idx + 1}. ", end="")
        print_card(card)

    choice = int(input("\nChoose a card to defend (1-5): ")) - 1
    player_card = player_deck[choice]
 
 # Player Attack > Opponent Health ---> Player win
    if opponent_card['Attack'] > player_card['Health']:
        print(f"\n{opponent_card['Name']} defeated your {player_card['Name']}!")
        player_deck.remove(player_card)
    else:
# caculation rules
        damage = opponent_card['Attack'] - player_card['Defense']
        if damage > 0:
            player_card['Health'] -= damage
            player_card['Defense'] = 0
        else:
            player_card['Defense'] -= opponent_card['Attack']

        print(f"\nAfter opponent's attack:")
        print_card(player_card)

    time.sleep(3)

while player_deck and opponent_deck:
    player_turn()
    if not opponent_deck:
        print("\nYou won the game!")
        break

    opponent_turn()
    if not player_deck:
        print("\nThe opponent won the game!")
        break

    print("\nNew round will begin soon...")
    time.sleep(3)


```
